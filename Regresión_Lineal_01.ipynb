````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
````

Genera datos aleatorios de ventas para un periodo de 180 días. Supondremos que las ventas tienen una tendencia lineal con algunas variaciones aleatorias.

````python
# Crear un conjunto de datos de 180 días
np.random.seed(42)  # Para reproducibilidad
dias = np.arange(1, 181).reshape(-1, 1)  # Período de tiempo en días

# Generar datos de ventas con una tendencia lineal y ruido aleatorio
ventas = 100 + 0.5 * dias + np.random.normal(0, 10, size=(180, 1))

# Visualizar los datos generados
plt.scatter(dias, ventas, color='blue')
plt.title('Ventas de Yoghurt vs Días')
plt.xlabel('Días')
plt.ylabel('Ventas')
plt.show()
````

Utiliza el modelo de regresión lineal de scikit-learn para ajustar una línea a los datos generados.

````python
# Crear el modelo de regresión lineal
modelo = LinearRegression()

# Entrenar el modelo con los datos
modelo.fit(dias, ventas)

# Obtener los parámetros de la regresión
pendiente = modelo.coef_[0][0]
interseccion = modelo.intercept_[0]

print(f'Pendiente: {pendiente}')
print(f'Intersección: {interseccion}')
````

Usa el modelo entrenado para predecir las ventas en el mismo periodo de 180 días.

````python
# Realizar predicciones
predicciones = modelo.predict(dias)

# Graficar las predicciones junto con los datos originales
plt.scatter(dias, ventas, color='blue', label='Datos reales')
plt.plot(dias, predicciones, color='red', label='Línea de regresión')
plt.title('Ventas de Yoghurt vs Días')
plt.xlabel('Días')
plt.ylabel('Ventas')
plt.legend()
plt.show()
````

Calcula el coeficiente de determinación 𝑅
2
 para evaluar la precisión del modelo

````python
# Evaluar el modelo con R^2
r2_score = modelo.score(dias, ventas)
print(f'Coeficiente de determinación R^2: {r2_score:.2f}')
````

````python
# Generar los próximos 180 días
nuevos_dias = np.arange(181, 361).reshape(-1, 1)

# Hacer predicciones para los próximos 180 días
predicciones_nuevos_dias = modelo.predict(nuevos_dias)

# Visualizar las predicciones junto con los datos originales
plt.scatter(dias, ventas, color='blue', label='Datos reales')
plt.plot(dias, predicciones, color='red', label='Línea de regresión (datos originales)')
plt.plot(nuevos_dias, predicciones_nuevos_dias, color='green', linestyle='dashed', label='Predicción futuros 180 días')
plt.title('Ventas de Yogurt: Predicciones para los próximos 180 días')
plt.xlabel('Días')
plt.ylabel('Ventas')
plt.legend()
plt.show()
````
